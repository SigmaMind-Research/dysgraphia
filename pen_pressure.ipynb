```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('temp.jpg')
img=cv2.resize(img,(900,900))
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(img,cmap='gray')
```

```python
(h,w)=np.shape(img)
```

```python
inverted=255-img
```

```python
plt.imshow(inverted,cmap='gray')
```

```python
bilateral = cv2.bilateralFilter(inverted, 5, 50, 50) 
```

```python
plt.imshow(bilateral,cmap='gray')
```

```python

ret,img=cv2.threshold(bilateral,110,255,cv2.THRESH_TOZERO)
plt.imshow(img,cmap='gray')
#print(np.unique(img))
```

```python
element=np.ones((3,3),np.uint8)
skel=np.zeros(np.shape(img),np.uint8)
original_variance=np.var(img)
```

```python
while True:
    open=cv2.erode(img,element)
    temp=img-open
    eroded=cv2.erode(img,element)
    skel=np.bitwise_or(skel,temp)
    img=eroded
   # print('a')
    if(np.all(img==0)):
        break
    
thinned=np.var(skel)
pres=original_variance-thinned

```
